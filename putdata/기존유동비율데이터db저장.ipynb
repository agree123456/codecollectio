```python
import numpy as np
import pandas as pd
import re
from datetime import datetime
from sqlalchemy import create_engine, text
import pandas as pd
```

```python
# 유동주식데이터 가져오기

data2_df = pd.read_csv('./20240105기준회사정보2.csv',header=[0,1,2])
data2_df = data2_df.drop(data2_df.iloc[0].name)
# 현재 인덱스를 'DatetimeIndex'로 변환
data2_df.index = pd.to_datetime(data2_df.index)
# 날짜 형식을 'YYYYMMDD'로 변경
data2_df.index = pd.to_datetime(data2_df.index, format='%Y%m%d')
df1 = data2_df.T.reset_index()
df2 = df1.drop(['level_0','level_2'],axis=1)
df3 = df2.T
```

```python
#new_df라는 빈 리스트를 생성합니다.
new_df = []
#range() 함수를 사용하여 df3 DataFrame의 열을 4개씩 선택합니다.
for i  in range(0, len(df3.columns), 4):
#선택한 열을 가진 DataFrame을 생성하고, df라는 변수에 저장합니다.
    df= (df3.iloc[1:, i:i+4])
# reset_index() 함수를 사용하여 DataFrame의 인덱스를 초기화합니다.
    df.reset_index(inplace=True)
# astype() 함수를 사용하여 'index' 열의 데이터 타입을 문자열로 변경합니다.
    df = df.astype({'index': str})
# '종목코드' 열을 생성하고, df3 DataFrame의 첫 번째 행에서 종목코드를 가져와 저장합니다.
    df['종목코드'] = np.array([df3.iloc[0,i] for k in range(733)])
# DataFrame의 열 순서를 조정합니다.
    col1 = [df.columns[-1]]
    col2 = df.columns[:-1].to_list()
    new_cols = col1+col2
    df1 = df[new_cols]
# dropna() 함수를 사용하여 결측값이 있는 행을 제거합니다.
    df1 =df1.dropna()
# replace() 함수를 사용하여 ',' 문자를 빈 문자열로 대체합니다.
    df1 = df1.replace(',', '')
# 열 이름을 'shrtk', 'shrdate', 'listedstock', 'floatingstock', 'floatingratio', 'marketcap'으로 변경합니다.
    df1 = df1.rename(columns={df1.columns[0]: 'shrtk'})
    df1 = df1.rename(columns={df1.columns[1]: 'shrdate'})
# 'shrdate' 열의 값을 datetime 형식으로 변환합니다.
    df1['shrdate'] = df1['shrdate'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
    df1 = df1.rename(columns={df1.columns[2]: 'listedstock'})
# 'listedstock' 열의 값에서 ',' 문자를 제거하고, 문자열로 변환합니다.
    df1['listedstock'] = df1['listedstock'].str.replace(',', '')
    df1['listedstock'] = df1['listedstock'].astype(str)
# 'floatingstock' 열의 값에서 ',' 문자를 제거하고, 실수형으로 변환합니다.
    df1 = df1.rename(columns={df1.columns[3]: 'floatingstock'})
    df1['floatingstock'] = df1['floatingstock'].str.replace(',', '')
    df1['floatingstock'] = df1['floatingstock'].astype(float)
# 'floatingratio' 열의 값을 정수형으로 변환합니다.
    df1 = df1.rename(columns={df1.columns[4]: 'floatingratio'})
    df1['floatingratio'] = df1['floatingratio'].astype(int)
# 'marketcap' 열의 값에서 ',' 문자를 제거하고, 정수형으로 변환합니다.
    df1 = df1.rename(columns={df1.columns[5]: 'marketcap'})
    df1['marketcap'] = df1['marketcap'].str.replace(',', '')
    df1['marketcap'] = df1['marketcap'].astype(int)
# 마지막으로, 생성된 DataFrame을 new_df 리스트에 추가합니다.
    new_df.append(df1)
```

```python
# create_engine() 함수를 사용하여 MariaDB에 연결합니다.
engine = create_engine("mysql+pymysql://big17:1234@34.64.63.23:3306/tradings?charset=utf8mb4")
conn = engine.connect()
# conn.execute() 함수를 사용하여 'SHAREHOLD1' 테이블에서 최대 'shseq' 값을 조회하고, 이를 'next_shseq' 변수에 저장합니다.
result = conn.execute(text("SELECT COALESCE(MAX(S.shseq), 0) + 1 AS next_shseq FROM SHAREHOLD1 S"))
next_shseq = result.scalar()
# new_df 리스트를 순회하면서 각 DataFrame을 삽입합니다.
for i, newnew_df in enumerate(new_df):
    # print(f'{i}번째 데이터프레임을 삽입합니다.')
    new_seq = []
    for idx, row in newnew_df.iterrows():
# 각 DataFrame의 첫 번째 열에 'shseq' 값을 생성합니다.
        new_seq.append(next_shseq + idx)
    newnew_df.insert(0,'shseq', new_seq)
    next_shseq += len(newnew_df)
# 'shseq' 값을 추가한 DataFrame을 'SHAREHOLD1' 테이블에 'append' 옵션으로 삽입합니다.
# 해당 코드를 실행하면, MariaDB에 연결하여 new_df 리스트의 각 DataFrame을 'SHAREHOLD1' 테이블에 삽입합니다.
    newnew_df.to_sql('SHAREHOLD1', conn, if_exists='append', index=False)
# conn.commit() 함수를 사용하여 데이터베이스에 변경 사항을 커밋합니다.
    conn.commit()
    print(i)
# 마지막으로, 연결을 종료합니다.
conn.close()
```
