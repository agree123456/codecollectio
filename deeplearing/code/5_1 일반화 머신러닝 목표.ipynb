```python
import numpy as np 
import pandas as pd 
import tensorflow
import warnings

from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.datasets import mnist
warnings.filterwarnings('ignore')
```

```python
(train_data , train_labels),_ = mnist.load_data() 
```

```python
train_data.shape
```

```python
train_images = train_data.reshape((60000,28*28))
train_images.shape
```

```python
train_images[0]
```

```python
train_labels[0]
```

```python
# 0-1사이 값으로 변환
train_images = train_images.astype('float32')/255
train_images[0]
```

```python
# np.random.randn/(len(train_images),784)[0]
np.random.random(((len(train_images),784)))[0]
```

```python
train_images_with_noise_channels = np.concatenate(
    [train_images,np.random.random((len(train_images),784))],axis=1
)
train_images_with_zeros_channels = np.concatenate(
    [train_images,np.zeros((len(train_images),784))],axis=1
)
```

```python
train_images_with_noise_channels.shape,train_images_with_zeros_channels.shape
```

```python
def get_model():
    model = keras.Sequential([
        layers.Dense(512,activation = 'relu'),
        layers.Dense(10,activation = 'softmax')   
    ])
    model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
    return model
```

```python
model =get_model()
history_noise = model.fit(
    train_images_with_noise_channels,
    train_labels,
    epochs = 10,
    batch_size = 128,
    validation_split = 0.2
)
```
