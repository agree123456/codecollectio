```python
# 간단한예제
inputs = keras.Input(shape = (3,),name = 'my_input')# input,output
weights = layers.Dense(64,activation = 'relu')(inputs)
outputs = layers.Dense(10,activation = 'softmax')(weights)
# 아직모델을 생성하지 않앗다.
model = keras.Model(inputs = inputs,outputs= outputs) # 다중입력과 다중출력이 가능한다.  # 리스트로 쌓아서 가능하다.
```

```python
# 다중입력과 다중출력
# 주가정보뉴스처리 
# 카테고리
voca_size = 10000
num_tags = 100
num_departiments = 4 # 4개부서 
# 다중입력 처리
title = keras.Input(shape=(voca_size,),name = 'title_Input')
text_body = keras.Input(shape=(voca_size),name='text_body')
tags = keras.Input(shape=(num_tags),name='tags')
# 웨이트를 의해 학습시키는것        
# 신경망 구성
features = layers.Concatenate()([title,text_body,tags])
features = layers.Dense(64,activation =  'relu')(features)

# 다중출력층 처리
priority = layers.Dense(1,activation = 'sigmoid',name = 'priority')(features)
department  =layers.Dense(num_departiments,activation  = 'softmax',name = 'department')(features)

# model 생성
model = keras.Model(inputs = [title,text_body,tags],outputs = [priority,department])
```

```python
model.summary()
```

```python
# 영향도를 컨트롤해봐라
```

```python
# 다중입력, 다중출력 모델 훈련
# 데이터를 임의의 값 답도 임의의 값
# 전체데이터 개수
num_samples = 1280
# 데이터 생성
# v np.random.randint(low , high = 숫자를 넣은거 미만 , size = None dtype = Int) == 디폴트
title_data = np.random.randint(0,2,size=[num_samples,voca_size])
text_body_data  = np.random.randint(0,2,size=[num_samples,voca_size]) 
tag_data = np.random.randint(0,2,size=[num_samples,num_tags])
# 답 생성
priority_label = np.random.random(size=(num_samples,1))
department_label = np.random.randint(0,2,size=[num_samples,num_departiments])
# 컴파일
model.compile( 
            optimizer = 'rmsprop'
            ,loss = ['mean_squared_error'
            ,"categorical_crossentropy"]
            ,metrics = [['mean_absolute_error'],["accuracy"]])
# 학습
model.fit(
    [title_data,text_body_data,tag_data]
    ,[priority_label,department_label]
    ,epochs = 1
)
# 평가
model.evaluate(
    [title_data,text_body_data,tag_data]
    ,[priority_label,department_label]
    
)
# 예측
priority_preds , department_preds  = model.predict(
    [title_data,text_body_data,tag_data]
)
```

```python
class CustomerTicketModel(keras.Model):

    def __init__(self, num_departments):
        super().__init__()
        self.concat_layer = layers.Concatenate()
        self.mixing_layer = layers.Dense(64, activation="relu")
        self.priority_scorer = layers.Dense(1, activation="sigmoid")
        self.department_classifier = layers.Dense(
            num_departments, activation="softmax")

    def call(self, inputs):
        title = inputs["title"]
        text_body = inputs["text_body"]
        tags = inputs["tags"]

        features = self.concat_layer([title, text_body, tags])
        features = self.mixing_layer(features)
        priority = self.priority_scorer(features)
        department = self.department_classifier(features)
        return priority, department
```

```python
model = CustomerTicketModel(num_departments=4)

priority, department = model(
    {"title": title_data, "text_body": text_body_data, "tags": tag_data})
```

```python
model.compile(optimizer="rmsprop",
              loss=["mean_squared_error", "categorical_crossentropy"],
              metrics=[["mean_absolute_error"], ["accuracy"]])
model.fit({"title": title_data,
           "text_body": text_body_data,
           "tags": tag_data},
          [priority_label, department_label],
          epochs=1)
model.evaluate({"title": title_data,
                "text_body": text_body_data,
                "tags": tag_data},
               [priority_label, department_label])
priority_preds, department_preds = model.predict({"title": title_data,
                                                  "text_body": text_body_data,
                                                  "tags": tag_data})
```
